```python
import numpy as np
import pandas as pd

from sklearn.tree import DecisionTreeClassifier
from sklearn.base import BaseEstimator
from sklearn.metrics import accuracy_score

from matplotlib import pyplot as plt
import seaborn as sns

import imageio
from IPython.display import Image, display
```

```python
class AdaBoost(BaseEstimator):
    def __init__(self, clfs=None, alphas=None, limit=55):
        self.clfs = clfs
        self.alphas = alphas
        self.limit = limit


    def fit(self, x_train, y_train):
        self.clfs = []
        self.alphas = []
        ws = [1 / len(x_train) for _ in x_train.iterrows()]

        for i in range(self.limit):
            clf = DecisionTreeClassifier(max_depth=2).fit(x_train, y_train, sample_weight=ws)
            predicted = [y_true == y_pred for y_true, y_pred in zip(y_train, clf.predict(x_train))]
            
            error = np.dot(np.logical_not(predicted).astype(int), ws)
            alpha = np.log((1 - error) / error) / 2

            ws = [w * np.exp(-alpha * (1 if p else -1)) for w, p in zip(ws, predicted)]
            ws /= sum(ws)
            
            self.clfs.append(clf)
            self.alphas.append(alpha)

        return self


    def predict(self, x_test):
        all_predictions = np.transpose(list(map(lambda clf: clf.predict(x_test), self.clfs)))
        return [np.sign(np.dot(self.alphas, list(map(int, predictions)))) for predictions in all_predictions]
    
    
    def __getitem__(self, range):
        return AdaBoost(self.clfs[range], self.alphas[range])
```

```python
def plot(df, clf, colorway='afmhot', name=''):
    x_min = df['x'].min()
    x_max = df['x'].max()
    y_min = df['y'].min()
    y_max = df['y'].max()
    x_back = np.ogrid[x_min:x_max:15j]
    y_back = np.ogrid[y_min:y_max:15j]
       
    space = [(x, y) for y in y_back for x in x_back]
    x_ax, y_ax = zip(*space)
    background = pd.DataFrame({'x': x_ax, 'y': y_ax, 'class': clf.predict(space)}) 
    z_back = background['class'].to_numpy().reshape(len(x_back), len(y_back))
    
    plt.figure(2)
    plt.contourf(x_back, y_back, z_back, cmap=colorway, alpha=0.8)
    plt.scatter(df['x'], df['y'], c=df['class'], cmap=colorway)
    plt.title(name)
    plt.savefig(f'{name}.png')
    plt.close(2)
```

```python
def run(df_name):
    df = pd.read_csv(f'{df_name}.csv')
    df.replace({'P': 1, 'N': -1}, inplace=True)
    x = df[['x', 'y']]
    y = df['class']
    
    clf = AdaBoost().fit(x, y)
    
    ks = range(1, 55 + 1)
    plt.plot(ks, list(map(lambda k: accuracy_score(clf[:k].predict(x), y), ks)))
    plt.xlabel('Size of committee')
    plt.ylabel('Accuracy on train');
    plt.show()

    name = lambda k: f'Iteration {k}'
    ks = [1, 2, 3, 5, 8, 13, 21, 34, 55]
    for k in ks:
        plot(df, clf[:k], name=name(k))

    images = []
    for pic in map(name, ks):
        for _ in range(3):
            images.append(imageio.imread(f'{pic}.png'))
    imageio.mimsave(f'{df_name}.gif', images)
```

```python
run('chips')
Image(data=open('chips.gif','rb').read())
```

```python
run('geyser')
Image(data=open('geyser.gif','rb').read())
```
